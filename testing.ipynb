## One Hot Encoding and Whatnot, punc removal

```python
# df = pd.read_csv("train.csv")

df = pd.DataFrame({
    "id": [1, 2, 3],
    "text": ["Hello! world.", "Machine learning is fun?", "Python, is great for data science"],
    "label": ["MWS", "EAP", "HPL"]
})

df.head()
```

```python
import re
from sklearn.feature_extraction.text import CountVectorizer

labels = df["label"]
df = df.drop(columns=["id", "label"])

# Remove punctuation
to_remove = ['.', ',', '?', '!', ':', ';']
replacement = " "

pattern = "|".join(map(re.escape, to_remove))
df["text"] = df["text"].str.replace(pattern, replacement, regex=True)

# Tokenization
vectorizer = CountVectorizer(lowercase=True, binary=True) # Binary makes values 0/1
X = vectorizer.fit_transform(df["text"]).toarray()

onehot_df = pd.DataFrame(X, columns=vectorizer.get_feature_names_out())

df_onehot = pd.concat([onehot_df, labels], axis=0)

print(labels)
print(onehot_df)
```
